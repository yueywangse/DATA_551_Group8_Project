```python
import pandas as pd

# Load file
df = pd.read_csv("crimedata_csv_AllNeighbourhoods_AllYears.csv")

# Ensure Year is numeric (safe conversion)
df["YEAR"] = pd.to_numeric(df["YEAR"], errors="coerce")

# Filter 2019–2023
filtered_df = df[df["YEAR"].between(2019, 2023)]

# Print count
print(filtered_df["YEAR"].value_counts().sort_index())
print("Number of records (2019–2023):", len(filtered_df))

# Save to file
filtered_df.to_csv("crimedata_2019_2023.csv", index=False)
```
